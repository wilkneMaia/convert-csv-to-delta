```python
import pandas as pd


def mostrar_cabecalho_parquet_pandas(parquet_path: str, n: int = 5):
    """
    Mostra o cabeçalho de um arquivo Parquet usando pandas.

    Args:
        parquet_path (str): Caminho do arquivo Parquet.
        n (int): Número de linhas a exibir.
    """
    df = pd.read_parquet(parquet_path)
    print("Colunas:", df.columns.tolist())
    print(df.head(n))


if __name__ == "__main__":
    parquet_path = "/Users/wilkne/Development/parquet-query/src/data/delta/part-00000-9b2a41c4-2b3d-46c0-8517-0f9ad9d4c39b-c000.snappy.parquet"
    mostrar_cabecalho_parquet_pandas(parquet_path)
```

```python
from pyspark.sql import SparkSession


def mostrar_cabecalho_parquet(parquet_path: str, n: int = 5):
    """
    Mostra o cabeçalho (nomes das colunas e primeiras linhas) de um arquivo Parquet usando PySpark.

    Args:
        parquet_path (str): Caminho do arquivo ou diretório Parquet.
        n (int): Número de linhas a exibir (default=5).
    """
    spark = SparkSession.builder.appName("MostrarCabecalhoParquet").getOrCreate()
    df = spark.read.parquet(parquet_path)
    print("Colunas:", df.columns)
    df.show(n)
    spark.stop()


if __name__ == "__main__":
    parquet_path = "/Users/wilkne/Development/parquet-query/src/data/parquet/data_2.parquet"
    mostrar_cabecalho_parquet(parquet_path)
```
